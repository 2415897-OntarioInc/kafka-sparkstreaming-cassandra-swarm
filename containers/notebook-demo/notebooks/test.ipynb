```python
import os
os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages org.apache.spark:spark-streaming-kafka-0-8_2.11:2.0.0,com.datastax.spark:spark-cassandra-connector_2.11:2.0.0-M3 pyspark-shell'
import time

from pyspark import SparkContext, SparkConf
from pyspark.sql import SQLContext, Row

SPARK_MASTER="local[2]"
SPARK_MASTER="spark://master:7077"
```

```python
#os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages com.databricks:spark-csv_2.10:1.3.0 pyspark-shell'
```

```python
conf = SparkConf() \
    .setAppName("Streaming test") \
    .setMaster(SPARK_MASTER) \
    .set("spark.cassandra.connection.host", "127.0.0.1") \
    .set("spark.ui.port","4040")
sc = SparkContext(conf=conf) 
sqlContext=SQLContext(sc)


```

```python
from pyspark.streaming import StreamingContext
from pyspark.streaming.kafka import KafkaUtils
```

```python
ssc = StreamingContext(sc, 1)
kvs = KafkaUtils.createStream(ssc, "kafka:2181", "spark-streaming-consumer", {'test': 1})
kvs.pprint()
```

```python
ssc.start()
```

```python
ssc.stop()
```
